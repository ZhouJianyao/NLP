# 3D

```python
import plotly.graph_objects as go
import pandas as pd
```

```python
data = pd.read_csv('./data/3d-line1.csv')
data.head()
```

```python
line = go.Scatter(x = data['x'], y = data['y'], mode='markers')
fig = go.Figure(line)
fig.show()
```

```python
line = go.Scatter3d(x = data['x'], y = data['y'], z = data['z'])
fig = go.Figure(line)
fig.show()
```

```python
line = go.Scatter3d(x = data['x'], y = data['y'], z = data['z'], mode='markers')
fig = go.Figure(line)
fig.show()
```

```python
line = go.Scatter3d(x = data['x'], y = data['y'], z = data['z'], 
                    mode='markers', marker={'size': 3, 'color': 'red'})
fig = go.Figure(line)
fig.show()
```
