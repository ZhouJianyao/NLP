```python
import jieba
import fasttext
```

```python
#读取停用词
stop_words = []
with open("data/project_data/stop_words.txt", "r", encoding="utf-8") as f_reader:
    for line in f_reader:
        line = line.replace("\r","").replace("\n","").strip()
        stop_words.append(line)
print(len(stop_words))
stop_words = set(stop_words)
print(len(stop_words))
```

```python
category = []
f_writer = open("data/example_data/train.txt", "w", encoding="utf-8")
with open("data/news.train.txt", "r", encoding="utf-8") as f_reader:
    for line in f_reader:
        line = line.replace("\r", "").replace("\n", "").strip()
        line_list = line.split("\t")
        if len(line_list) == 2:
            seg_list = jieba.cut(line_list[1])
            word_list = []
            for word in seg_list:
                if word not in stop_words:
                    word_list.append(word)
            line = " ".join(word_list)
            line = "__label__" + line_list[0] + "\t" + line + "\n"
            f_writer.write(line)
            f_writer.flush()
            if line_list[0] not in category:
                category.append(line_list[0])
f_writer.close
```

```python
f_writer = open("data/example_data/test.txt", "w", encoding="utf-8")
with open("data/news.test.txt", "r", encoding="utf-8") as f_reader:
    for line in f_reader:
        line = line.replace("\r", "").replace("\n", "").strip()
        line_list = line.split("\t")
        if len(line_list) == 2:
            seg_list = jieba.cut(line_list[1])
            word_list = []
            for word in seg_list:
                if word not in stop_words:
                    word_list.append(word)
            line = " ".join(word_list)
            line = "__label__" + line_list[0] + "\t" + line + "\n"
            f_writer.write(line)
            f_writer.flush()
f_writer.close
```

```python
print(category)
```

```python
#文本分类训练，输入路径，输出model路径
model = fasttext.supervised("data/train.txt", "classifier.model")
```

```python
train_result = model.test("data/train.txt")
#准确率
print("准确率：", train_result.precision)
print("召回率：", train_result.recall)
```

```python
test_result = model.test("data/test.txt")
```

```python
print("test准确率：", test_result.precision)
print("test召回率：", test_result.recall)
```
