```python
import pandas as pd
pd.set_option('display.max_columns', None)
path = './'
train = pd.read_csv(path + 'input/pfm_train.csv')
test = pd.read_csv(path + 'input/pfm_test.csv')
```

```python
test['Attrition'] = -1
data = train.append(test).reset_index(drop=True)
```

```python
data.drop(['Over18', 'StandardHours'], axis=1, inplace=True)
```

```python
cat_col = [i for i in data.select_dtypes(object).columns if i not in ['Attrition']]
```

```python
def encode_onehot(df, column_name):
    feature_df = pd.get_dummies(df[column_name], prefix=column_name)
    concat = pd.concat([df.drop([column_name], axis=1), feature_df], axis=1)
    return concat
```

```python
for i in cat_col:
    data = encode_onehot(data, i)
```

```python
data.EmployeeNumber.value_counts()
```

```python
data.drop(['EmployeeNumber'], axis=1, inplace=True)
# 删除唯一值
```

```python
data.head()
```

```python
feats = [i for i in data.columns if i not in ['Attrition']]
```

```python
import numpy as np

max_min_scaler = lambda x: (x - np.min(x)) / (np.max(x) - np.min(x))
for i in feats:
    data[i] = data[[i]].apply(max_min_scaler)
```

```python
data.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression

train = data[data['Attrition'] != -1][feats]
train = train.fillna(0)
label = data[data['Attrition'] != -1]['Attrition']
test = data[data['Attrition'] == -1][feats]
test = test.fillna(0)

model = LogisticRegression()
model.fit(train, label)
```

#### 直接得到单模结果

```python
p = model.predict_proba(test)[:,1]
test['result'] = p
test[['result']].to_csv(path + 'output/lr_proba.csv', index=False)
```

```python
del test['result']
y_hat = model.predict(test)
test['result'] = y_hat
test[['result']].to_csv(path + 'output/lr_submission.csv', index=False)
```

使用网络搜索选择最优参数

```python
from sklearn.model_selection import GridSearchCV
lg = LogisticRegression()
clf = GridSearchCV(lg, param_grid=[{'C':np.arange(0.001, 0.05, 0.001)}], cv=5)
clf.fit(train, label)
best_model = clf.best_estimator_
clf.best_params_
```

```python
del test['result']
result = best_model.predict(test)
pd.DataFrame({'result':result}).to_csv(path + 'output/grid_lrsubmission.csv', index=False)
```
